```python
numero = int(input("Digite um número de 1 a 7: "))

dias = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

if 1 <= numero <= 7:
    print("Dia da semana:", dias[numero])
else:
    print("Número inválido!")
```

```python
horas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

if horas <= 40:
    salario = horas * valor_hora
else:
    extras = horas - 40
    salario = (40 * valor_hora) + (extras * valor_hora * 1.5)

print("Salário total: R$", salario)
```
